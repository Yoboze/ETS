```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
file_path = 'New_cleaned_dataset.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Convert boolean columns for country groups to 0 and 1
boolean_columns = [
    'COUNTRY_GROUPED_DE', 'COUNTRY_GROUPED_FR', 'COUNTRY_GROUPED_IN',
    'COUNTRY_GROUPED_Others', 'COUNTRY_GROUPED_US'
]

for col in boolean_columns:
    data[col] = data[col].astype(int)

# Display the first few rows to verify the changes
data[boolean_columns].head()
```

```python
# Ensure the dataset retains its original column count of 24 columns
# Check the current number of columns
initial_columns = data.columns.tolist()

# Convert boolean columns to 0 and 1
for col in boolean_columns:
    data[col] = data[col].astype(int)

# Ensure the dataset retains the same number of columns
current_columns = data.columns.tolist()

# Check if any columns have been added or removed
if len(initial_columns) != len(current_columns):
    raise ValueError("Column count has changed during processing.")

# Display the first few rows to verify the changes
data.head()
```

```python
# Save the updated dataset to a new CSV file
updated_file_path = 'Updated_New_cleaned_dataset.csv'
data.to_csv(updated_file_path, index=False)

updated_file_path
```
